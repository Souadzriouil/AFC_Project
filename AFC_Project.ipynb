```python
import pandas as pd
from fanalysis.ca import CA
%matplotlib inline
import numpy
import matplotlib.pyplot as plt
```

```python
# Chargement des données depuis un fichier Excel
data = pd.read_excel("C:\AD\donnees_nettoyees_World_Cup_.xlsx", sheet_name=0, index_col=0)
print(data)                  
```

```python
from scipy.stats import chi2_contingency

# Utilisation de la fonction chi2_contingency
chi2, pvalue, degrees, expected = chi2_contingency(data)

# Affichage des résultats
print("chi2 =", chi2)
print("degrés de liberté =", degrees)
print("p-valeur =", pvalue)
```

```python
#importation de la librairie
from fanalysis.ca import CA
#lancer les calculs
afc = CA(row_labels=data.index,col_labels=data.columns)
afc.fit(data.values)
print(dir(afc))
```

```python
#Affichage des valeur propres
print(afc.eig_)
```

```python
afc.plot_eigenvalues()
```

```python
# Coordonnées des points lignes
print(afc.row_coord_)
```

```python
#Contribution des points lignes
print(afc.row_contrib_)
```

```python
#Coordonnes des points colonnes
print(afc.col_coord_)
```

```python
#Contribution des points colonnes
print(afc.col_contrib_)
```

```python
#Cos2 des point colonnes
print(afc.col_cos2_)
```
